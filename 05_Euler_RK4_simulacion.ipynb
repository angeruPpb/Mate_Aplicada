```python
import numpy as np
import math
from matplotlib import pyplot as plt

def f(x, y):
    return (-0.9906/-1000)*y*(1000-y)

def f_y(x):
    return 1000/(1+999*math.exp(-0.9906*x))

def error(vect_mse):
    vect_h = [0.1, 0.05, 0.025]
    plt.figure(figsize=(5, 3))
    plt.plot(vect_h, vect_mse, marker='o')
    plt.title("Error / h")
    plt.xlabel("h")
    plt.ylabel("Mean Squared Error")
    plt.show()
```

```python
# Euler normal
def euler_method(x=0, y=0, h=0.1):
    x_values = []
    y_values = []
    r_values = []
    mse_values = []
    # print("x\t y\t Real\t eAbs\t %eRel")

    while x <= 10.1:
        r = f_y(x)
        eAbs = r-y
        eRel = eAbs/r * 100
        # print("{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}".format(x,y,r,eAbs,eRel))
        x_values.append(x)
        y_values.append(y)
        r_values.append(r)
        mse_values.append(eAbs**2)
        y = y + h * f(x, y)
        x = x + h
    plt.figure(figsize=(5, 3))
    plt.plot(x_values, y_values, label='Euler Values')
    plt.plot(x_values, r_values, label='Real Values')
    plt.legend()
    plt.title(h)
    plt.show()
    mse = np.mean(mse_values)
    print("MSE({:.4f}):{:.4f}".format(h,mse))
    return mse
```

```python
# Euler x=0 y=1
mse_1 = euler_method(x=0,y=1,h=0.1)
mse_2 = euler_method(x=0,y=1,h=0.05)
mse_3 = euler_method(x=0,y=1,h=0.025)

vect_mse = [mse_1, mse_2, mse_3]
error(vect_mse)
```

```python
# RK4
def rugen_kuta_4(x=0,y=0,h=0.1):
    x_values = []
    y_values = []
    r_values = []
    mse_values = []

    while x < 10.1:
        k1 = f(x,y)
        k2 = f(x+h/2,y+(h/2)*k1)
        k3 = f(x+h/2,y+(h/2)*k2)
        k4 = f(x+h,y+h*k3)
        r = f_y(x)
        eAbs = r-y
        eRel = eAbs/r * 100
        x_values.append(x)
        y_values.append(y)
        r_values.append(r)
        mse_values.append(eAbs**2)
        y = y+(h/6)*(k1+2*k2+2*k3+k4)
        x = x+h

    plt.figure(figsize=(5, 3))
    plt.plot(x_values, y_values, label='RK4 Values')
    plt.plot(x_values, r_values, label='Real Values')
    plt.legend()
    plt.title(h)
    plt.show()

    mse = np.mean(mse_values)
    print("MSE({:.4f}):{:.15f}".format(h,mse))
    return mse
```

```python
# Rugen Kuta x=0 y=1
mse_1 = rugen_kuta_4(x=0,y=1,h=0.1)
mse_2 = rugen_kuta_4(x=0,y=1,h=0.05)
mse_3 = rugen_kuta_4(x=0,y=1,h=0.025)

vect_mse = [mse_1, mse_2, mse_3]
error(vect_mse)
```
