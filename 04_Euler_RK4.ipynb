```python
import numpy as np
import math
from matplotlib import pyplot as plt

# y' = f(x,y) = 2xy
def f(x, y):
    return 2*x*y

def f_y(x):
    return math.exp((x**2)-1)
```

```python
# Euler normal
def euler_method(x=0, y=0, h=0.1):
    x_values = []
    y_values = []
    print("x\t y\t Real\t eAbs\t %eRel")

    while x <= 1.5:
        r = f_y(x)
        eAbs = r-y
        eRel = eAbs/r * 100
        print("{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}".format(x,y,r,eAbs,eRel))
        x_values.append(x)
        y_values.append(y)
        y = y + h * f(x, y)
        x = x + h

    plt.plot(x_values, y_values)
    plt.title(h)
    plt.show()
```

```python
euler_method(x=1,y=1,h=0.1)
euler_method(x=1,y=1,h=0.05)
euler_method(x=1,y=1,h=0.025)
```

```python
# Euler Mejorado
def euler_method_plus(x=0, y=0, h=0.1):
    x_values = []
    y_values = []

    print("x\t y\t Real\t eAbs\t %eRel")

    while x < 1.51:
        y1 = y+h*f(x,y)
        r = f_y(x)
        eAbs = r-y
        eRel = eAbs/r * 100
        print("{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}".format(x,y,r,eAbs,eRel))
        x_values.append(x)
        y_values.append(y)
        y = y + h * (f(x, y)+f(x+h,y1))/2
        x = x + h


    plt.plot(x_values, y_values)
    plt.title(h)
    plt.show()
```

```python
euler_method_plus(x=1,y=1,h=0.1)
euler_method_plus(x=1,y=1,h=0.05)
euler_method_plus(x=1,y=1,h=0.025)
```

```python
# RK2
def rugen_kuta_2(x=0,y=0,h=0.1):
    x_values = []
    y_values = []
    print("x\t y\t Real\t eAbs\t %eRel")

    while x < 1.51:
        k1 = f(x,y)
        k2 = f(x+h,y+h*k1)
        r = f_y(x)
        eAbs = r-y
        eRel = eAbs/r * 100
        print("{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}".format(x,y,r,eAbs,eRel))
        x_values.append(x)
        y_values.append(y)
        y = y+(h/2)*(k1+k2)
        x = x+h

    plt.plot(x_values, y_values)
    plt.title(h)
    plt.show()
```

```python
rugen_kuta_2(x=1,y=1,h=0.1)
rugen_kuta_2(x=1,y=1,h=0.05)
rugen_kuta_2(x=1,y=1,h=0.025)
```

```python
# RK4
def rugen_kuta_4(x=0,y=0,h=0.1):
    x_values = []
    y_values = []
    print("x\t y\t Real\t eAbs\t %eRel")

    while x < 1.51:
        k1 = f(x,y)
        k2 = f(x+h/2,y+(h/2)*k1)
        k3 = f(x+h/2,y+(h/2)*k2)
        k4 = f(x+h,y+h*k3)
        r = f_y(x)
        eAbs = r-y
        eRel = eAbs/r * 100
        print("{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}\t{:.4f}".format(x,y,r,eAbs,eRel))
        x_values.append(x)
        y_values.append(y)
        y = y+(h/6)*(k1+2*k2+2*k3+k4)
        x = x+h

    plt.plot(x_values, y_values)
    plt.title(h)
    plt.show()
```

```python
rugen_kuta_4(x=1,y=1,h=0.1)
rugen_kuta_4(x=1,y=1,h=0.05)
rugen_kuta_4(x=1,y=1,h=0.025)
```
